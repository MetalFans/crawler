```python
def getOpenHours(urls):
    openList = []
    header = {
        'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.94 Safari/537.36'
    }
    ck = {
        '__flash_buy_ad_cookie':'ok'
    }
    rs = (grequests.get(url, cookies=ck, headers=header) for url in urls)
    res = grequests.map(rs, size=15)
    for r in res:
        soup = bs(r.text,'html.parser')
        dict = {}
        open = soup.select('.hours')
        if open[0].select('tr'):
            for o in open[0].select('tr'):
                dict[o.select('td')[0].text] = " / ".join(o.select('td')[1].text.split())
        else:
            dict[u'營業時間'] = open[0].text.split()[1].replace(u'，',' / ')
        openList.append(dict)
    return openList
```

```python
# 各篇文章分別跑
import re
import requests
import time
import math
import numpy as np
import operator
from bs4 import BeautifulSoup as bs
from collections import Counter
import os

if not os.path.exists('./Workspace/data/dictionary'):
    os.makedirs('./Workspace/data/dictionary')
if not os.path.isfile('./Workspace/data/dictionary/autodict_v2.txt'):
    with open('./Workspace/data/dictionary/autodict_v2.txt', 'w'):
        pass
    
def removePunctuation(source):
    xx = u"([^0-9a-zA-Z\u4e00-\u9fff]+)"
    s = re.sub(xx,' ',source)
    return s

def removeTag(soup,tag):
    [x.extract() for x in soup.select(tag)]
    
def entropy(n_gram_lengths, text):
    entropies = {}
    for n in n_gram_lengths:
        length_n_substrings = [text[i:i+n] for i in range(len(text)-n+1)]
        length_n_substring_counts = Counter(length_n_substrings)
        normalizer = float(len(length_n_substrings))
        distribution_values = np.array(length_n_substring_counts.values()) / normalizer
        entropy = -np.sum(p*np.log2(p) for p in distribution_values)
        entropies[n] = entropy
    return entropies

def getAll(section, page):
    domain = 'http://www.appledaily.com.tw'
    res = requests.get('http://www.appledaily.com.tw/realtimenews/section/%s/%d' % (section, page))
    soup = bs(res.text, 'html.parser')
    urls = [domain+t['href'] for t in soup.select('.rtddt a')]
    return getContent(urls)

def getContent(urls):
    news = []
    for u in urls:
        res = requests.get(u)
        soup = bs(res.text,'html.parser')
        apple = soup.select('#summary')[0].text
        news.append(removePunctuation(apple))
    return news

for p in xrange(1,11):
    news = getAll('strange',p)
    for tx in news:
        sp = tx.split()
        for n in xrange(2,3):
            substring_counts_dict = {}
            for s in sp:
                sentence_substrings = [s[i:i+n] for i in range(len(s)-n+1)]
                for word in sentence_substrings:
                    if word not in substring_counts_dict:
                        substring_counts_dict[word] = 1
                    else:
                        substring_counts_dict[word] += 1            
            length_count_dict = dict((k,v) for k,v in substring_counts_dict.items())
            word_freq = sorted(length_count_dict.iteritems(),key=operator.itemgetter(1),reverse=True)
            org = entropy([n], tx).values()[0]
            word_entropy = {}
            word_freq = [w for w in word_freq if w[1]>=2 and len(re.sub('\w','',w[0])) == 2]
            for w in word_freq:
                word_entropy[w[0]] = entropy([n], tx.replace(w[0],'')).values()[0]
                print w[0], w[1]
            weighted_world = sorted(word_entropy.iteritems(),key=operator.itemgetter(1),reverse=False)
            for keyword in weighted_world:
                if (org-keyword[1])/org >= 0.002:
                    with open('./Workspace/data/dictionary/autodict_v2.txt', 'a') as f:
                        f.write('%s %s %s' % (keyword[0].encode('utf-8'), str((org-keyword[1])/org), '\n'))
```
